Script para modificar los labels de los datasets de roboflow para su unificación

```python
# dronefinalyear.v1i.yolov11
import os

mapping = {
    0: 6,   # Animal
    1: 7,   # Casa 
    2: 9,   # Arbustos
    3: 8,   # Arbol
    4: 0    # Coche
}

root = "dronefinalyear.v1i.yolov11"

for split in ["train", "valid", "test"]:
    folder = os.path.join(root, split, "labels")
    if not os.path.exists(folder):
        continue
    for file in os.listdir(folder):
        if not file.endswith(".txt"):
            continue
        path = os.path.join(folder, file)
        lines = open(path).read().strip().splitlines()
        new_lines = []
        for line in lines:
            cls, x, y, w, h = line.split()
            cls_new = mapping[int(cls)]
            new_lines.append(f"{cls_new} {x} {y} {w} {h}\n")
        open(path, "w").writelines(new_lines)
```
